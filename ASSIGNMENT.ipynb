```python
merged_df = pd.merge(details_df, orders_df, on='Order ID')
```

```python
merged_df
```

```python
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'], format='%d-%m-%Y')
```

## Total Sales

```python
total_sales = merged_df['Amount'].sum()
print(f"Total Sales: ₹{total_sales}")
```

## Average Sales per Region

```python
avg_sales_per_region = merged_df.groupby('State')['Amount'].mean()
print("\nAverage Sales per Region:")
print(avg_sales_per_region)
```

## Sales by Product Category

```python
sales_by_category = merged_df.groupby('Category')['Amount'].sum()
print("\nSales by Product Category:")
print(sales_by_category)
```

## Monthly Sales Trends

```python
merged_df['Month'] = merged_df['Order Date'].dt.to_period('M')
monthly_sales_trends = merged_df.groupby('Month')['Amount'].sum()
print("\nMonthly Sales Trends:")
print(monthly_sales_trends)
```

```python
sales_by_geography = merged_df.groupby(['State', 'City'])['Amount'].sum()
print("\nSales Distribution by Geography:")
print(sales_by_geography)
```
